```python
windgustdir=pd.get_dummies(rain_data['WindGustDir'], drop_first=True)
```

```python
rain_data.shape
```

```python
winddir9am = pd.get_dummies(rain_data['WindDir9am'], drop_first=True)
winddir9am.head(5)
```

```python
winddir3pm = pd.get_dummies(rain_data['WindDir3pm'], drop_first=True)
winddir3pm.head(5)
```

```python
location = pd.get_dummies(rain_data['Location'], drop_first=True)
location.head(5)
```

```python
rain_data['Cloud3pm'].fillna((rain_data['Cloud3pm'].mean()), inplace=True)
```

```python
rain_data['Cloud9am'].fillna((rain_data['Cloud9am'].mean()), inplace=True)
```

```python
rain_data['WindSpeed9am'].fillna((rain_data['WindSpeed9am'].mean()), inplace=True)
```

```python
rain_data['WindSpeed3pm'].fillna((rain_data['WindSpeed3pm'].mean()), inplace=True)
```

```python
rain_data['Rainfall'].fillna((rain_data['Rainfall'].mean()), inplace=True)
```

```python
rain_data['MinTemp'].fillna((rain_data['MinTemp'].mean()), inplace=True)
```

```python
rain_data['MaxTemp'].fillna((rain_data['MaxTemp'].mean()), inplace=True)
```

```python
rain_data['WindGustSpeed'].fillna((rain_data['MaxTemp'].mean()), inplace=True)
```

```python
rain_data['Humidity9am'].fillna((rain_data['Humidity9am'].mean()), inplace=True)
```

```python
rain_data['Humidity3pm'].fillna((rain_data['Humidity3pm'].mean()), inplace=True)
```

```python
rain_data['Pressure9am'].fillna((rain_data['Pressure9am'].mean()), inplace=True)
```

```python
rain_data['Pressure3pm'].fillna((rain_data['Pressure3pm'].mean()), inplace=True)
```

```python
rain_data['Temp3pm'].fillna((rain_data['Temp3pm'].mean()), inplace=True)
```

```python
rain_data['Temp9am'].fillna((rain_data['Temp9am'].mean()), inplace=True)
```

```python
rain_data['RISK_MM'].fillna((rain_data['RISK_MM'].mean()), inplace=True)
```

```python
rain_data=pd.concat([rain_data, windgustdir, winddir9am, winddir3pm, location], axis=1)
rain_data.head(5)
```

```python
rain_data.isnull().sum()
```

```python
rain_data.drop("RainToday",axis=1,inplace=True)
```

```python
rain_data.drop(['WindGustDir','WindDir9am','WindDir3pm','Location','Date'],axis=1,inplace=True)
```

```python
X=rain_data.drop('RainTomorrow',axis=1)
y=rain_data['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.isnull().sum()
```

```python
X_train.head(6)
```

```python
X_train.isnull().sum() > 0
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
