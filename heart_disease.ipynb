```python
X = df.drop("HeartDisease", axis=1)
y = df["HeartDisease"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.40,random_state=42)
```

```python
df = df.replace("Male",1)
df = df.replace("Female",0)
```

```python
X_train = X_train.replace("Male",1)
X_train = X_train.replace("Female",0)

X_test = X_test.replace("Male",1)
X_test = X_test.replace("Female",0)

y_train = y_train.replace("Male",1)
y_train = y_train.replace("Female",0)

y_test = y_test.replace("Male",1)
y_test = y_test.replace("Female",0)
```

```python
X_train = X_train.drop("AgeCategory", axis=1)
X_test = X_test.drop("AgeCategory", axis=1)
```

```python
X_train = X_train.drop("Race", axis=1)
X_test = X_test.drop("Race", axis=1)
```

```python
X_train = X_train.drop("GenHealth", axis=1)
X_test = X_test.drop("GenHealth", axis=1)
```

```python
X_train = X_train.replace("No, borderline diabetes",0.5)
X_test = X_test.replace("No, borderline diabetes",0.5)
```

```python
X_train = X_train.replace("Yes (during pregnancy)",1)
X_test = X_test.replace("Yes (during pregnancy)",1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg_pred = log_reg.predict(X_test)
log_reg_acc = metrics.accuracy_score(y_test, log_reg_pred)
print("logistic regression model accuracy(in %):", log_reg_acc*100)
```

```python
X_train.head(20)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(max_leaf_nodes=3, random_state=0, criterion='entropy')
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
dt_acc = metrics.accuracy_score(y_test, dt_pred)

print('Decision Tree Classifier accuracy (in%): ', dt_acc*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn_clf = KNeighborsClassifier(n_neighbors=6)
kn_clf.fit(X_train, y_train)
kn_pred = kn_clf.predict(X_test)
kn_acc = metrics.accuracy_score(y_test, kn_pred)

print('KNeighborsClassifier accuracy (in%): ',kn_acc*100)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gradientboost_clf = GradientBoostingClassifier(max_depth=2, random_state=1)
gradientboost_clf.fit(X_train,y_train)
gradientboost_pred = gradientboost_clf.predict(X_test)
gradientboost_acc = metrics.accuracy_score(y_test, gradientboost_pred)

print('GradientBoostingClassifier accuracy (in%): ',gradientboost_acc*100)
```
